# Scraping 100 Articles 

```python
from bs4 import BeautifulSoup
import requests
import csv
import pandas as pd
```

```python
linkList = []

for i in range(15):
    source = requests.get('https://www.detik.com/search/searchall?query=korupsi&siteid=2&sortby=time&page='+str(i)).text
    
    soup = BeautifulSoup(source, 'lxml')
    
    for article in soup.find_all('article'):
        
        link = article.a['href']
        
        if "news.detik.com" in link:
            linkList.append(link)
    
with open('scrape.csv', 'w') as f:
    j = 0
    for item in linkList:
        print(j)
        print(item)
        f.write("%s\n" % item)
        j = j + 1
```

```python
df = pd.read_csv('scrape.csv', header=None)
df
```

```python
import io
```

```python
articleList = []
for i in range(100):
    link = df[0][i]
    source = requests.get(link).text
    
    soup = BeautifulSoup(source, 'lxml')

    article = soup.findAll('p')

    for a in soup.findAll('a'):
        a.replaceWithChildren()
    
    word = ''
    for j in range(len(article)-1):
        word = word + article[j].get_text()
        
    articleList.append(word)

with io.open('article.txt', 'w', encoding="utf-8") as f:
    j = 0
    for item in articleList:
        f.write("%s\n" % item)
        j = j + 1
```

# Bigram Model

```python
import nltk
nltk.download('punkt')
```

```python
f = open("article.txt", "r")

sentenceList = []

for i in f:
    sentenceList.append(i.lower())
```

```python
import re
```

```python
words = str(sentenceList).lower()
words = re.sub('[^A-Za-z0-9 ]+', '', words)
tokens = nltk.word_tokenize(words)
```

```python
tokens
```

```python
len(tokens)
```

```python
freq_tab = {}
total_count = 0
for token in tokens:
    if token in freq_tab:
        freq_tab[token] += 1
    else:
        freq_tab[token] = 1
    total_count += 1
        
print(freq_tab)
```

```python
prob_tab = {}
for token in freq_tab:
    prob_tab[token] = freq_tab[token]/total_count
    
print(prob_tab)
```

```python
bigram_freq_tab = {}
sents = str(words).split('\n')
for sent in sents:
    sent_tokens = nltk.word_tokenize(sent)
    sent_tokens_length = len(sent_tokens)
    for i in range(0,sent_tokens_length-1): 
        strkey = sent_tokens[i]+','+sent_tokens[i+1]
        if strkey in bigram_freq_tab:
            bigram_freq_tab[strkey] += 1
        else:
            bigram_freq_tab[strkey] = 1 
    print('----')  
print(bigram_freq_tab)
```

```python
prob_tab = {}

for par_tokens in bigram_freq_tab:
    after = par_tokens.split(',')
    prob_tab[par_tokens] = bigram_freq_tab[par_tokens]/freq_tab[after[1]]
    
print(prob_tab)
```

Test Sentence (closely related to the topic)

```python
test_sentence = 'Pemerintah terus melanjutkan proyek yang belum tuntas'
lc_test_sentence = test_sentence.lower()

sents = str(lc_test_sentence).split(' ')
test_tokens = []
for i in range(0, len(sents)-1):
    test = sents[i]+','+sents[i+1]
    if test in prob_tab:
        test_tokens.append(prob_tab[test])
    else:
        test_tokens.append(1/len(tokens))
        
total_prob = 1.0
for test_token in test_tokens:
    total_prob = total_prob * test_token
    
print(total_prob)
```

```python
perplexity = (1/total_prob)**(1/(len(test_tokens)+1))
print(perplexity)
```
